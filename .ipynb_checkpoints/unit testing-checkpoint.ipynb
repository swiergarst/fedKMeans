## "federated" implementation classes

```python
class beun_client():
    def __init__(self, data, n_centroids):
        self.data = data
        self.n_centroids = n_centroids
        self.means = np.zeros((n_centroids, 2))
    def predict_means(self, global_centers):
        
        local_centers = self.det_new_local_centers(global_centers)
        
        km = KMeans(n_clusters = self.n_centroids, init = local_centers, max_iter = 1).fit(self.data)
        mean_vals = np.copy(km.cluster_centers_)
        self.means = np.copy(km.cluster_centers_)
        samples = self.calc_cluster_amts(km)
        return mean_vals, km.score(self.data), samples
                                
    def calc_cluster_amts(self, km):
        return np.array([len(km.labels_[km.labels_ == i]) for i in range(self.n_centroids)])
        
    def det_new_local_centers(self, global_centers):
        #simple option: calculate distances from data mean
        dmean = np.mean(self.data, axis = 0)
        dists = np.linalg.norm(global_centers - dmean, axis = 1)
        dist_inds = np.argsort(dists)
        
        gcenter_mat = np.array([global_centers for i in range(self.n_centroids)])
        lcenter_mat = np.array([self.means for i in range(global_centers.shape[0])])
        #print(gcenter_mat.shape, lcenter_mat.shape)
        return global_centers[dist_inds[0:self.n_centroids]]
        # slightly more complicated option: calculate distances between all center-pairs, 
        # then select the closest ones per previous mean

        '''
        distances = np.zeros((self.n_centroids, global_centers.shape[0]))
        for i in range(self.n_centroids):
            distances[i,:] = np.linalg.norm(global_centers - self.means[i], axis = 0 )
        '''    
        
```

```python
class beun_server():
    def __init__(self, n_clusters):
        self.n_clusters = n_clusters
            
    def aggregate(self, means, samples):
        
        cluster_aggregator = KMeans(n_clusters = 16)
        avg_means = np.zeros((self.n_clusters, 2))
        
        means_res = means.reshape((-1,2))
        #cluster_aggregator.fit(means_res, sample_weight =  samples.reshape(-1))
        cluster_aggregator.fit(means_res, sample_weight = samples)

        
        return cluster_aggregator.cluster_centers_, 0
                         
```

## full run

```python
def calc_cluster_distance(local_clusters):
    # output: a n-1 by n-1 array of distances (could make it n by n with diagonal 0)
    a = np.zeros((local_clusters.shape[0], local_clusters.shape[0]))
    # this is super unoptimized but w/e
    for i in range(local_clusters.shape[0]):
        for j in range(local_clusters.shape[0]):
            a[i,j] = np.linalg.norm(local_clusters[i,:] -  local_clusters[j,:])
    return a
```

```python
def det_local_clusters(clients, n_cluster_g = 16, alpha = 0.5, xmin = -12.5, xmax = 12.5, ymin = -12.5, ymax = 12.5):
    nclients = len(clients)
    #clients = [beun_client(np.genfromtxt(datafiles[i], delimiter=','), n_cluster_g) for i in range(nclients)]


    pre_threshold = math.sqrt(((xmin - xmax) * (ymin - ymax))) / n_cluster_g
    thr = pre_threshold * alpha


    local_clusters = np.zeros((nclients, n_cluster_g, 2))
    local_cluster_amts = np.array([n_cluster_g] * nclients)

    for j in range(n_cluster_g - 1):
        for i, client in enumerate(clients):
            clusters_l, _  = kmeans_plusplus(client.data, local_cluster_amts[i])

            a = calc_cluster_distance(clusters_l)
            if (a<thr).sum() > local_cluster_amts[i]:
                local_cluster_amts[i] -= 1
    return local_cluster_amts
```

```python
def run(crounds, nclients, n_cluster_global, alpha):
    clients = [beun_client(np.genfromtxt(datafiles[i], delimiter=','), n_cluster_global) for i in range(nclients)]
    server = beun_server(n_cluster_global)

    local_cluster_amts = det_local_clusters(clients, alpha = alpha)
    print(local_cluster_amts)

    for client_i, client in enumerate(clients):
        client.n_centroids = local_cluster_amts[client_i]


    mean_log = np.zeros((n_cluster_global,2, crounds))
    mean_log_client = np.zeros((n_cluster_global,2,crounds,nclients))
    weight_log = np.zeros((n_cluster_global * nclients, crounds))
    scores = np.zeros((nclients, crounds))


    #init_clusters = np.zeros((n_cluster_global, 2, nclients))
    for i, client in enumerate(clients):
        init_clusters_local, _ = kmeans_plusplus(client.data, local_cluster_amts[i])
        if i == 0:
            init_clusters = init_clusters_local
        else:
            init_clusters = np.concatenate((init_clusters, init_clusters_local))

    #print(init_clusters.shape)
    init_km = KMeans(n_clusters = n_cluster_global).fit(init_clusters)

    cluster_centers = np.copy(init_km.cluster_centers_)
    cluster_begin = np.copy(cluster_centers)
    for r in range(crounds):

        for i, client in enumerate(clients):
            mean,  scores[i,r], sample = client.predict_means(cluster_centers)
            if i == 0:
                means = mean
                samples = sample
            else:
                means = np.concatenate((means, mean), axis = 0)
                samples = np.concatenate((samples, sample), axis = 0)
            #print("centers ", str(i), ": ", cluster_centers)
        #mean_log_client[:,:,r,:] = means
            #print(samples)
        cluster_centers, weight_log[:,r] = server.aggregate(means, samples)
        mean_log[:,:,r] = cluster_centers
            
    #for client in clients:
    #    client.set_new_means(new_means)
    return mean_log, cluster_centers, scores
    
```

## other paper implementation

```python
class observer():
    def __init__(self, data, n_centroids):
        self.data = data
        self.n_centroids = n_centroids
        self.kmeans = KMeans(n_clusters = n_centroids)
        self.kmeans.fit(self.data)
        self.cluster_amts  = np.array([self.kmeans.labels_[self.kmeans.labels_ == i].sum() for i in range(self.n_centroids)])
        
    def determine_bin_samples(self, bin_x, bin_y, bin_size): #there's probably a super fancy python way of doing this
        bin_amount = 0
        for data_i in range(self.data.shape[0]):
            point_x = self.data[data_i, 0]
            point_y = self.data[data_i, 1]
            if (point_x > bin_x) & (point_x < (bin_x + bin_size)):
                if (point_y > bin_y) & (point_y < (bin_y + bin_size)):
                    bin_amount += 1
        return bin_amount
    def determine_bin_amt(self, bin_x, bin_y, bin_size): # the fancy try
        data_adjusted = np.array([self.data[:,0] - bin_x, self.data[:,1] - bin_y])
        data_filtered = data_adjusted[:, np.all(data_adjusted>0, axis = 0)]
        data_final = data_filtered[:, np.all(data_filtered<bin_size, axis=0)]
        return int(data_final.shape[1])
    
    def determine_bins(self,min_x, min_y, max_x, max_y, bin_size): # we keep square bins for now
        bins = np.zeros((math.ceil((max_x - min_x)/bin_size), math.ceil((max_y - min_y)/bin_size)), dtype=int)
        total_bins_x = bins.shape[0]
        total_bins_y = bins.shape[1]
        print(total_bins_x, total_bins_y)
        for b_x in range(total_bins_x):
            for b_y in range(total_bins_y):
                bins[b_y,b_x] = int(self.determine_bin_amt(min_x + bin_size * b_x, min_y + bin_size * b_y, bin_size))
        return bins
        
    def fit(self, means):
        self.kmeans.cluster_centers_ = np.copy(means)
        new_labels = self.kmeans.predict(self.data)
        return new_labels
        
    def compute_gradients(self):
        g = np.zeros((self.n_centroids, 2))
        for point_i in range(self.data.shape[0]):
            label = self.kmeans.labels_[point_i]
            g[label, :] += self.kmeans.cluster_centers_[label, :] - self.data[point_i, :]
        return g
```

```python
class central_server():
    def __init__(self, n_centroids, alpha, x_min, y_min, x_max, y_max):
        self.z = np.zeros((n_centroids, 2))
        self.m = np.zeros_like(self.z)
        self.alpha = alpha
        self.x_min = x_min
        self.y_min = y_min
        self.x_max = x_max
        self.y_max = y_max
        #self.bin_size = bin_size
        
    def calc_w(self, s_list):
        total_samples = np.zeros(s_list.shape[1])
        for s in range(s_list.shape[0]):
            total_samples += s_list[s,:]
            
        ws = np.zeros_like(s_list)    

        ws = np.copy(s_list) / total_samples
        self.ws = np.nan_to_num(ws, nan=1/ s_list.shape[0])
        return ws

    def calc_z(self, g_list):
        for i, g in enumerate(g_list):
            
            self.z[:, 0] +=  self.ws[i,:] * g[:,0]
            self.z[:, 1] += self.ws[i,:] * g[:,1]
    
    def calc_m(self):
        self.m = (-1 * np.copy(self.z))/ (2 * self.alpha)
```
